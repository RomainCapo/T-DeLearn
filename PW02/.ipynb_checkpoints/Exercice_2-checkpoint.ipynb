# Implementation of sigmoid function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1.0 / (1.0+np.exp(-z))

def derivative_sigmoid(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
z_vec = np.linspace(-10,10,100)
```

```python
plt.plot(z_vec,sigmoid(z_vec),'k-',label="sigmoid")
plt.plot(z_vec,derivative_sigmoid(z_vec),'k--',label="derivative")
plt.legend()
plt.show()
```
